```python
import os, json, yaml
import random, shutil
from pathlib import Path
from PIL import Image

while ".git" not in os.listdir(os.getcwd()): prj_dir = os.chdir(Path(os.getcwd())/"..")

from utils.boxes import xywh2rxywh
from utils.others import createEmptyFolders

def i2id(i): 
    return str(i).zfill(5)

base_folder = Path("data/unified/yolo-sets")
```

```python
name = "self-svd"

train_folder = base_folder/name/"train"
val_folder = base_folder/name/"val"

createEmptyFolders([train_folder, val_folder])
```

```python
with open(f'data/unified/{name}.json', 'r') as file:
    data = json.load(file)
```

```python
base = data 
# base = list(filter(lambda x: x['source'] == "svd", data))
# base = base[:10]

for i, e in enumerate(base):
    file_name_base = i2id(i)

    src_file_name = e['file_name']
    trg_file_name = train_folder/f"{file_name_base}.jpg"

    im = Image.open(src_file_name)
    print(f"Image {file_name_base}: {im.size}")

    if Path(src_file_name).suffix == '.jpg':
        print("  copy ...")
        shutil.copyfile(src_file_name, str(trg_file_name))
    else:
        print("  convert ...")
        im = Image.open(src_file_name)
        im.save(str(trg_file_name))


    with open(f"{train_folder}/{file_name_base}.txt", 'w') as f:
        for a in e['annotations']:
            cid = 0
            bbox = xywh2rxywh(a['bbox'], im.size)

            f.write(f"{cid} {' '.join([str(round(t, 8)) for t in bbox])}\n")
```

```python
iss = list(range(len(base)))
random.shuffle(iss)

for i in range(round(len(iss)/20)):
    file_name_base = i2id(iss[i])
    Path(train_folder/f"{file_name_base}.jpg").rename(val_folder/f"{file_name_base}.jpg")
    Path(train_folder/f"{file_name_base}.txt").rename(val_folder/f"{file_name_base}.txt")
```

### Write YAML meta

```python
with open(base_folder/name/"data.yaml", 'w') as file:
    yaml.dump({
           'nc': 1,
        'names': ['vehicle'],
        'train': './train/',
          'val': './val/',
    }, file, default_flow_style=False)
```
